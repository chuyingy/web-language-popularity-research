```python
topics_2017 ={}
for i in range(len(github_2017)):
    for topic in ast.literal_eval(github_2017.iloc[i]['topics']):
        if topic not in topics_2017.keys():
            topics_2017[topic] = 1
        else:
            topics_2017[topic] += 1

topics_2018 ={}
for i in range(len(github_2018)):
    for topic in ast.literal_eval(github_2018.iloc[i]['topics']):
        if topic not in topics_2018.keys():
            topics_2018[topic] = 1
        else:
            topics_2018[topic] += 1

topics_2019 ={}
for i in range(len(github_2019)):
    for topic in ast.literal_eval(github_2019.iloc[i]['topics']):
        if topic not in topics_2019.keys():
            topics_2019[topic] = 1
        else:
            topics_2019[topic] += 1

topics_2020 ={}
for i in range(len(github_2020)):
    for topic in ast.literal_eval(github_2020.iloc[i]['topics']):
        if topic not in topics_2020.keys():
            topics_2020[topic] = 1
        else:
            topics_2020[topic] += 1

topics_2021 ={}
for i in range(len(github_2021)):
    for topic in ast.literal_eval(github_2021.iloc[i]['topics']):
        if topic not in topics_2021.keys():
            topics_2021[topic] = 1
        else:
            topics_2021[topic] += 1

topics_2022 ={}
for i in range(len(github_2022)):
    for topic in ast.literal_eval(github_2022.iloc[i]['topics']):
        if topic not in topics_2022.keys():
            topics_2022[topic] = 1
        else:
            topics_2022[topic] += 1
```
