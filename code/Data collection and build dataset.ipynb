```python
import numpy as np
import math
import requests
import json
import pandas as pd
```

### Crawl data from Github

```python
# insert your own account name and password
auth_name = ''
auth_pwd = ''
```

```python
query_url = 'https://api.github.com/search/repositories?q=topic:web+pushed:2017-01-01..2017-03-31&per_page=100&page=1'
```

```python
response_result = requests.get(query_url, auth=(auth_name, auth_pwd))
response_result.json()
```

```python
for year in [2017,2018,2019,2020,2021,2022]:
    months = ["01-01","03-31","04-01","06-30","07-01","09-30","10-01","12-31"]
    for month_index in [0,2,4,6]:
        date_start = months[month_index]
        date_end = months[month_index+1]
        total_page = 1
        page = 1
        while (page <= total_page): 
            query_string = 'https://api.github.com/search/repositories?q=topic:web+pushed:{}-{}..{}-{}&per_page=100&page={}'.format(year, date_start, year, date_end, page)
            response = requests.get(query_string, auth=(auth_name, auth_pwd))
            data = response.json()
            if (page == 1):
                total_page = math.ceil(data['total_count']/100)
                with open('data/github_{}_{}.json'.format(year,date_start), 'w') as outfile:
                    json.dump(data, outfile)
                outfile.close()
            else:
                with open('data/github_{}_{}.json'.format(year,date_start)) as json_file:
                    store_result = json.load(json_file)
                json_file.close()
                store_result['items'].extend(data['items'])
                with open('data/github_{}_{}.json'.format(year,date_start), 'w') as outfile:
                    json.dump(store_result, outfile)
                outfile.close()
            page = page + 1
    
```

### build dataset

#### 2017

```python
with open('data/github_2017_01-01.json') as json_file:
    github_2017_01 = json.load(json_file)
json_file.close()

with open('data/github_2017_04-01.json') as json_file:
    github_2017_04 = json.load(json_file)
json_file.close()  

with open('data/github_2017_07-01.json') as json_file:
    github_2017_07 = json.load(json_file)
json_file.close()  

with open('data/github_2017_10-01.json') as json_file:
    github_2017_10 = json.load(json_file)
json_file.close()
```

```python
keys = github_2017_04['items'][0].keys()
keys
```

```python
for key in keys:
    print(key,end=', ')
```

```python
github_2017 = github_2017_01['items']+github_2017_04['items']+github_2017_07['items']+github_2017_10['items']
```
